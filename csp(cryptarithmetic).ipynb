```python
import itertools

def solve_crypto(word1, word2, result):
    letters = sorted(set(word1 + word2 + result))

    if len(letters) > 10:
        print("More than 10 unique letters not allowed")
        return []

    first_letters = {word1[0], word2[0], result[0]}
    solutions = []

    for perm in itertools.permutations(range(10), len(letters)):
        mapping = dict(zip(letters, perm))

        # Constraint: no leading zero
        if any(mapping[ch] == 0 for ch in first_letters):
            continue

        def value(word):
            return int("".join(str(mapping[c]) for c in word))

        if value(word1) + value(word2) == value(result):
            solutions.append(mapping)

    return solutions


def print_solutions(solutions):
    print("\n ALL VALID SOLUTIONS:\n")
    for i, sol in enumerate(solutions, 1):
        print(f"Solution {i}:")
        for k in sorted(sol):
            print(f"  {k} = {sol[k]}")
        print()


def evaluate_query(query, solutions):
    query = query.replace(" ", "")
    variables = query.split("+")

    print("\n QUERY EVALUATION FOR EACH SOLUTION:\n")
    for i, sol in enumerate(solutions, 1):
        total = sum(sol[v] for v in variables)
        values = " + ".join(str(sol[v]) for v in variables)
        print(f"Solution {i}: {values} = {total}")


# ---------------- MAIN ----------------
if __name__ == "__main__":
    print(" Cryptoarithmetic CSP using Backtracking\n")

    word1 = input("Enter first word  : ").upper()
    word2 = input("Enter second word : ").upper()
    result = input("Enter result word : ").upper()

    solutions = solve_crypto(word1, word2, result)

    print(f"\nTotal number of valid solutions: {len(solutions)}")

    if not solutions:
        exit()

    print_solutions(solutions)

    query = input("Enter query (example: R+O+T): ").upper()
    evaluate_query(query, solutions)
```
